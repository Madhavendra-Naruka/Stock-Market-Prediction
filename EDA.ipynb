```python
import os
import pandas as pd
import numpy as np
```

```python
#Reading the Data
numeric_df = pd.read_csv("combined_stock_data.csv")

# Ensure the 'Date' column is of datetime type
numeric_df['Date'] = pd.to_datetime(numeric_df['Date'])
                                    
# Group by 'Date' and compute the mean for numeric columns
mean_df = numeric_df.groupby('Date').mean()


print(numeric_df.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 3))
# plt.plot(mean_df.index, mean_df['Open'], label='Open')
plt.plot( mean_df['High'], label='High')
plt.plot( mean_df['Low'], label='Low')
# plt.plot(mean_df.index, mean_df['Close'], label='Close')
# plt.plot(mean_df.index, mean_df['Adj Close'], label='Adj Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Mean Stock Prices Over Time')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 3))
plt.plot(mean_df.index, mean_df['Open'], label='Open')
# plt.plot( mean_df['High'], label='High')
# plt.plot( mean_df['Low'], label='Low')
plt.plot(mean_df.index, mean_df['Close'], label='Close')
# plt.plot(mean_df.index, mean_df['Adj Close'], label='Adj Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Mean Stock Prices Over Time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 3))
plt.plot( mean_df['Volume'], label='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Mean Trading Volume Over Time')
plt.legend()
plt.show()
```

```python
num_companies=86
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(mean_df['Close']/num_companies, label='Close Price')
plt.plot(mean_df['sma_50']/num_companies, label='SMA_50')
plt.plot(mean_df['sma_200']/num_companies, label='SMA_200')

plt.title(f"Mean Moving Averages ")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.subplot(2,1,2)
plt.plot(mean_df['Close']/num_companies, label='Close Price')
plt.plot(mean_df['ema_50']/num_companies, label='EMA_50')
plt.plot(mean_df['ema_200']/num_companies, label='EMA_200')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

plt.show()
plt.close()
```

```python
plt.figure(figsize=(12, 3))
plt.plot(mean_df.index, mean_df['rsi'], label='RSI')
plt.axhline(y=70, color='r', linestyle='--', label='Overbought')
plt.axhline(y=30, color='g', linestyle='--', label='Oversold')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Average RSI Over Time')
plt.legend()
plt.show()
```
